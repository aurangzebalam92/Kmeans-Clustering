```python
#load the libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
#Creating dummy data for clustering
from sklearn.datasets import make_blobs
```

```python
Data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.7, random_state = 101)
print(Data[0])
```

```python
Data[1]
```

```python
plt.scatter(Data[0][:,0],Data[0][:,1],c=Data[1],cmap="rainbow")
```

```python
#KMEANS
from sklearn.cluster import KMeans
```

```python
#Instantiating the cluster object
K_Means = KMeans()
K_Means.fit(Data[0])
```

```python
K_Means.cluster_centers_
```

```python
K_Means.labels_
```

```python
"""
Here our analysis stops in production.
Because this is supposed to be unsupervised Data and we will have no way to check the validity of the Data

"""
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,sharex=True,figsize=(10,6))
ax1.set_title("This is by K means")
ax1.scatter(Data[0][:,0],Data[0][:,1],c=K_Means.labels_,cmap="rainbow")
ax2.set_title("Original labels from make_blob")
ax2.scatter(Data[0][:,0],Data[0][:,1],c=Data[1],cmap="rainbow")
```

```python
mycenters = np.array(K_Means.cluster_centers_)
mycenters
```

```python
plt.scatter(mycenters[:,0],mycenters[:,1],marker = "X", linewidths = 30)
```
